Unix time (also known as **Epoch time**) is a system for describing a point in time.    
It is the number of seconds that have elapsed since the Unix epoch (minus leap seconds);

# Python Datetime with **date**, **time**, and **datetime**
### בפייטון יש לנו מודול מובנה בשביל להתמודד עם זמנים ותאריכים


#### איך יוצרים?

```python
from datetime import date, time, datetime
date(year=2020, month=1, day=31)
```

```python
time(hour=13, minute=14, second=31)
```

```python
datetime(year=2020, month=1, day=31, hour=13, minute=14, second=31)
```

#### מה היום הנוכחי?

```python
print(f'{date.today()} - Using date')
print(f'{datetime.now()} - Using datetime')
```

### ליצור אובייקט מטקסט

אם הטקסט לפי ISO 8601

```python
date.fromisoformat("2020-01-31") # לפי ISO 8601
```

אפשר לעשות פורמט לפי סימנים

```python
date_string = "01-31-2020 14:45:37"
format_string = "%m-%d-%Y %H:%M:%S"
datetime.strptime(date_string, format_string)
```

### How to Format

| Component | Code | Value |
| --- | --- | --- |
| Year (as four-digit integer ) | %Y | 2020
| Month (as zero-padded decimal) | %m | 01
| Date (as zero-padded decimal) | %d | 31
| Hour (as zero-padded decimal with 24-hour clock) | %H | 14
| Minute (as zero-padded decimal) | %M | 45
| Second (as zero-padded decimal) | %S | 37

### להציג את האובייקט זמן בצורה שמתאימה לנו

```python
now = datetime.now()
print(now.strftime("%d-%m-%Y"))
print(now.strftime("%d/%m/%Y"))
print(now.strftime("%H:%M"))
```

###  חיבור וחיסור זמנים

```python
from datetime import timedelta
print(now.strftime("%d-%m-%Y"))
next_week = now + timedelta(days=7)
print(f'Adding 7 days: {next_week.strftime("%d-%m-%Y")}')
```

### timedelta
